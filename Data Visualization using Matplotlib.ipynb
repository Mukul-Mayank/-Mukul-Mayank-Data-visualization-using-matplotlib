pie chart

```python
mx=np.random.randint(10,60,(8))
mx
```

```python
x=[10, 14, 15, 27, 10, 13, 53, 59]
y=["JavaScript","Java","C++","C","PHP","Swift","R","Python"]
col = ["Red", "Blue", "Green", "Yellow", "Orange", "Purple", "Pink", "Magenta"]
explode = [0, 0.1, 0, 0.2,0.7,0.1,0.3,0.4]
```

```python
plt.figure(figsize=(20,6))
plt.pie(x,labels=y,colors=col,autopct='%1.1f%%',shadow=True)
plt.title("pie-chart")
plt.legend(loc=3)
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.pie(x,labels=y,colors=col,explode=explode,autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
plt.figure(figsize=(15,4))
plt.pie(x,labels=y,colors=col,explode=explode,autopct='%1.1f%%',shadow=True,labeldistance=2)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.pie(x,labels=y,colors=col,explode=explode,autopct='%1.1f%%',shadow=True,labeldistance=2,startangle=90,counterclock=False)
plt.show()
```

Area v/s Stack plot

```python
#single stack plot
x=[41, 43, 32, 34, 43, 37, 17, 44, 18, 47, 47, 20, 23, 18, 32]
y=[24, 43, 16, 31, 28, 31, 26, 34, 58, 29, 56, 53, 24, 54, 34]
plt.stackplot(x,y)
plt.show()
```

```python
#multiple stack plot
x=[41, 43, 32, 34, 43, 37, 17, 44, 18, 47, 47, 20, 23, 18, 32]
y=[24, 43, 16, 31, 28, 31, 26, 34, 58, 29, 56, 53, 24, 54, 34]
z=[27, 20, 27, 31, 23, 25, 44, 36, 20, 41, 30, 38, 52, 21, 20]
p=[40, 26, 17, 32, 49, 25, 49, 58, 42, 27, 41, 58, 38, 39, 17]
plt.stackplot(x,y,z,p)
plt.grid()
plt.show()

```

Step Plot

```python
x=[41, 43, 32, 34, 43, 37, 17, 44, 18, 47, 47, 20, 23, 18, 32]
y=[24, 43, 16, 31, 28, 31, 26, 34, 58, 29, 56, 53, 24, 54, 34]
plt.step(x,y,color="r",marker="*",ms=10,mfc="g",label="step_plot")
plt.legend(loc=4)
plt.show()
```

Fill between plot

```python
x=[41, 43, 32, 34, 43, 37, 17, 44, 18, 47, 47, 20, 23, 18, 32]
y=[24, 43, 16, 31, 28, 31, 26, 34, 58, 29, 56, 53, 24, 54, 34]
plt.plot(x,y,color="r")
plt.fill_between(x,y,color="g",alpha=0.1)
plt.show()
```
